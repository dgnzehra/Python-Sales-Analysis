```python
import pandas as pd
import os
```

```python
# merge the 12 months of data into a single CSV file

files = [file for file in os.listdir('./SalesAnalysis/Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./SalesAnalysis/Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("all_data.csv", index=False)
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

### Clean up the data

```python
# Drop the raws of NaN

nan_df = all_data[all_data.isna().any(axis=1)]
all_data.head()

all_data = all_data.dropna(how="all")
all_data.head()
```

### Augment data with additional columns 

### Clean up the data 

```python
# Find 'Or' and delete

all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

```python
# Coverting columns to the correct type

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) #make float
```

### Adding a month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Adding a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Adding a city column

```python
# .apply() 

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) +' ('+ get_state(x) + ')')
all_data.head()
```

### What was the best month for sales? How much was earned that month? 

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1, 13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.xlabel('Month Number')
plt.ylabel('Sales in USD (1000000$)')
plt.show()
```

### What city had the highest number of sales?

```python
results = all_data.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.xlabel('City Name')
plt.ylabel('Sales in USD (1000000$)')
plt.show()
```

### What time should we display the advertisements to maximize likelihood of customer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()

# Around 11am and 7pm could be a good time to display the advertisements
```

### What products are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)].copy()

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int64')
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)


```

### What product sold the most? Why do you think it sold the most?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group['Quantity Ordered'].agg('sum')  # Explicitly specify 'sum' aggregation
products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.xticks(rotation='vertical')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'], errors='coerce')

all_data = all_data.dropna(subset=['Price Each'])

prices = all_data.groupby('Product')['Price Each'].mean()

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Qıuantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
